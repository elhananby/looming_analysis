# Looming analysis — quickstart

Load recordings, run the analysis pipeline, and render the standard publication plots.

**Edit the paths in the next cell, then run all cells.**

```python
from pathlib import Path

from looming_analysis import (
    AnalysisConfig,
    ResponsivenessConfig,
    find_braidz,
    run_analysis,
    plot_responses,
    plot_heading_traces,
    plot_heading_changes,
    plot_responsiveness_rates,
    plot_turn_proportions,
    plot_peak_velocity,
    plot_response_latency,
)
from looming_analysis.responsiveness import compute_turn_direction
```

```python
# --- Edit these ---
DATA_ROOT = Path("/mnt/data/experiments")

# Single group: all files in one directory
file_groups = {"experiment": find_braidz(DATA_ROOT)}

# Multiple groups (compare genotypes / conditions):
# file_groups = {
#     "CS":       find_braidz(DATA_ROOT / "CS"),
#     "DNp03":    find_braidz(DATA_ROOT / "DNp03"),
# }

print({g: len(f) for g, f in file_groups.items()}, "files")
```

```python
result = run_analysis(
    file_groups,
    analysis=AnalysisConfig(
        pre_ms=100.0,
        post_ms=500.0,
        cache_dir=".braidz_cache",
    ),
    responsiveness=ResponsivenessConfig(
        method="combined",
        threshold_deg_s=500.0,
        heading_threshold_deg=45.0,
        baseline_window_ms=(-90.0, -10.0),
        window_ms=(100.0, 200.0),
    ),
)

responses = result.responses
print(f"{len(responses)} trials extracted")
```

```python
# Scalar trial table (one row per trial)
trials = result.to_dataframe(kind="scalar")
trials.head()
```

```python
# Mean ± SD angular velocity traces
plot_responses(responses, col_by="stimulus_offset_deg", hue_by="group");
```

```python
# Mean heading traces
plot_heading_traces(responses, col_by="stimulus_offset_deg", hue_by="group");
```

```python
# Heading-change distribution
plot_heading_changes(responses, col_by="stimulus_offset_deg", hue_by="group");
```

```python
# Responsiveness rates per condition
plot_responsiveness_rates(responses, col_by="stimulus_offset_deg", hue_by="group");
```

```python
# Turn direction proportions
plot_turn_proportions(responses, col_by="stimulus_offset_deg", group_by="group");
```

```python
# Peak angular velocity distribution
plot_peak_velocity(responses, col_by="stimulus_offset_deg", hue_by="group");
```

```python
# Response latency
plot_response_latency(responses, hue_by="group");
```
